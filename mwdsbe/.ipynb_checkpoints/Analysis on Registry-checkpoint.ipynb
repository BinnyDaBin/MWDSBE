# Analysis on Registry

```python
import mwdsbe
import pandas as pd
import numpy as np
import geopandas as gpd
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
```

```python
registry = mwdsbe.load_registry() # geopandas df
```

## Neighborhoods with the most registered business

1. Convert registry to GeoDataframe
2. Import neighbordhoos from zillow
3. Combine these two

```python
registry_geo = registry.dropna(subset=['geometry'])
```

```python
len(registry_geo)
```

```python
registry_gdf = gpd.GeoDataFrame(registry_geo, geometry='geometry', crs={"init": "epsg:4326"})
```

```python
registry_gdf = registry_gdf.to_crs(epsg=3857)
```

```python
zillow = gpd.read_file('./data/neighborhood/zillow_neighborhoods.geojson')
zillow = zillow.to_crs(epsg=3857)
```

```python
registry = gpd.sjoin(registry_gdf, zillow, op='within', how='left')
```

```python
nbhoods = registry.groupby('ZillowName').size().to_frame('N').reset_index()
```

```python
top_10_nbhoods = nbhoods.sort_values(by='N', ascending=False)[:10]
```

```python
sns.set(style="whitegrid")
sns.set_palette(sns.cubehelix_palette(10))
ax = sns.barplot(x="N", y="ZillowName", data=top_10_nbhoods)
ax.set(xlabel='Number of Companies', ylabel='Neighborhoods')
plt.show()
```

```python
bottom_10_nbhoods = nbhoods.sort_values(by='N', ascending=False)[-10:]
```

```python
sns.set(style="whitegrid")
sns.set_palette(sns.cubehelix_palette(10))
ax = sns.barplot(x="N", y="ZillowName", data=bottom_10_nbhoods)
ax.set(xlabel='Number of Companies', ylabel='Neighborhoods')
plt.show()
```

```python
import altair as alt
alt.renderers.enable('notebook')
```

```python
nbhoods_geo = zillow.merge(nbhoods, on='ZillowName', how='left')
```

```python
nbhoods_geo['N_per_area'] = nbhoods_geo['N'] / (nbhoods_geo.geometry.area)
```

```python
# create an altair data source from a GeoJSON representation
data = alt.InlineData(
    values=nbhoods_geo.to_crs(epsg=4326).to_json(),
    format=alt.DataFormat(property="features", type="json"),
)

# plot map, where variables ares nested within `properties`,
alt.Chart(data).mark_geoshape(stroke="white").properties(
    width=500, height=400, projection={"type": "mercator"}
).encode(
    tooltip=["properties.ZillowName:N", "properties.N:N"],
    color=alt.Color("properties.N:Q", scale=alt.Scale(scheme='redpurple'))
)
```
