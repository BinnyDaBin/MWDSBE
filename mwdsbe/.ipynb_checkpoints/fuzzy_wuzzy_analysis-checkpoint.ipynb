## 1. Active companies by neighborhoods in Philadelphia
only considering active companies with geometry info in philly neighborhoods

```python
active_companies = matched[matched.license_status == 'Active']
print('Active Companies:', len(active_companies), 'out of', len(matched))
```

```python
active_companies_geo = active_companies.dropna(subset=['geometry'])
```

```python
print('Active Companies with geometry information:', len(active_companies_geo), 'out of', len(active_companies))
```

```python
# convert geometry from string to wkt
from shapely import wkt
active_companies_geo['geometry'] = active_companies_geo['geometry'].apply(wkt.loads)
```

```python
#convert matched dataframe to geodataframe
active_companies_gdf = gpd.GeoDataFrame(active_companies_geo, geometry='geometry', crs={"init": "epsg:4326"})
```

```python
active_companies_gdf = active_companies_gdf.to_crs(epsg=3857)
```

```python
# import neighborhood from zillow
zillow = gpd.read_file('data/neighborhood/zillow_neighborhoods.geojson')
zillow = zillow.to_crs(epsg=3857)
```

```python
joined_active_philly = gpd.sjoin(active_companies_gdf, zillow, op='within', how='left')
```

```python
active_philly = joined_active_philly.groupby('ZillowName').size()
active_philly = active_philly.to_frame('N').reset_index()
```

```python
active_philly = zillow.merge(active_philly, on='ZillowName', how='left').fillna(0)
```

### Barplot - Top 5 neighborhoods

### Interactivity

```python
import altair as alt
alt.renderers.enable('notebook')
```

```python
active_philly['N_per_area'] = active_philly['N'] / (active_philly.geometry.area)
```

```python
# create an altair data source from a GeoJSON representation
data = alt.InlineData(
    values=active_philly.to_crs(epsg=4326).to_json(),
    format=alt.DataFormat(property="features", type="json"),
)

# plot map, where variables ares nested within `properties`,
alt.Chart(data).mark_geoshape(stroke="white").properties(
    width=500, height=400, projection={"type": "mercator"}
).encode(
    tooltip=["properties.N_per_area:Q", "properties.ZillowName:N"],
    color=alt.Color("properties.N_per_area:Q", scale=alt.Scale(scheme='viridis'))
)
```

## 2. In-State Companies

```python
in_state_companies = active_companies[active_companies.location_state == 'PA']
print('Total number of active in-state Companies:', len(in_state_companies))
```

### Barchart

```python
states = active_companies.groupby(['location_state']).size()
states = states.to_frame('N').reset_index()
```

```python
in_state = states.location_state == 'PA'
```

```python
out_of_state = states[~in_state]
in_state = states[in_state]
```

```python
binary_states = in_state.append({'location_state': 'Out-of-State', 'N': out_of_state.sum().N}, ignore_index=True)
```

```python
binary_states.location_state.iloc[0] = 'In-State'
```

```python
binary_states
```

```python
ax = sns.barplot(x="location_state", y="N", data=binary_states)
ax.set(xlabel='Location', ylabel='Number of Companies')
plt.show()
```
