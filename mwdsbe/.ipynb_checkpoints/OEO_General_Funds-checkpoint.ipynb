# Compare OEO and General Funds

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import mwdsbe
import schuylkill as skool
```

```python
registry = mwdsbe.load_registry() # geopandas df
```

```python
gf = pd.read_excel(r'C:\Users\dabinlee\Documents\GitHub\mwdsbe_binny\MWDSBE\mwdsbe\data\cwedp_37_report.xlsx', sheet_name='general_funds')
```

```python
gf.head()
```

```python
len(gf)
```

```python
gf2 = gf.dropna(subset=['VEND_NAME'])
gf2 = gf2.loc[gf2['MAJ_CLASS'] != 1]
```

```python
gf2.head()
```

```python
len(gf2)
```

```python
prime_vendors = gf2.groupby('VEND_NAME')['AMT'].sum().to_frame('AMT').reset_index()
```

```python
top10_pv = prime_vendors.sort_values(by='AMT', ascending=False)[:10]
```

```python
top10_pv
```

```python
sns.set(style="whitegrid")
ax = sns.barplot(x="AMT", y="VEND_NAME", data=top10_pv)
ax.set(xlabel='Amount', ylabel='Vendor Names')
plt.show()
```
