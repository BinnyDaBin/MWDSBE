```python
import mwdsbe
```

```python
registry = mwdsbe.load_registry() # geopandas df
```

```python
registry.head()
```

```python
registry_geo = registry.dropna(subset=["geometry"])
```

```python
registry_geo
```

```python
registry.index
```

```python
registry_geo['id'] = registry_geo.index
```

```python
registry_geo.set_index('id')
```

```python
imp_registry = registry_geo[['id', 'company_name', 'geometry']]
```

```python
imp_registry.set_index('id', inplace=True)
```

```python
imp_registry.rename(columns={"company_name": "name"})
```

```python
imp_registry['type'] = "marker"
```

```python
imp_registry
```

```python
imp_registry.dropna(inplace=True)
```

```python
import pandas as pd
registry_df = pd.DataFrame(imp_registry)
```
